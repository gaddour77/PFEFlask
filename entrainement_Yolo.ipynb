```python
!pip install ultralytics
from ultralytics import YOLO

# Charger ton modèle déjà entraîné
model = YOLO("best.pt")

# Relancer l'entraînement avec des hyperparamètres optimisés
model.train(
    data="/content/MAE-car-damage-4/data.yaml",  # dataset
    epochs=100,          # plus d’epochs pour mieux converger
    imgsz=640,           # taille standard YOLO
    batch=16,            # ajuste selon ta VRAM (8 si GPU limité)
    pretrained=True,     # utilise best.pt comme point de départ
    optimizer="AdamW",   # meilleur que SGD dans beaucoup de cas
    lr0=0.001,           # learning rate initial (0.001 = bon point de départ)
    lrf=0.01,            # facteur de décroissance du LR
    momentum=0.937,      # momentum classique YOLO
    weight_decay=0.0005, # régularisation pour éviter l’overfitting
    patience=20,         # early stopping si pas d’amélioration
    augment=True,        # active data augmentation (par défaut: True)
    cos_lr=True          # scheduler "cosine" pour mieux converger
)

```
