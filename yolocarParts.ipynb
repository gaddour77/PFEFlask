```python
# ==========================================================
# 1. Installer Ultralytics (YOLOv8)
# ==========================================================
!pip install ultralytics

# Vérification
from ultralytics import YOLO


# ==========================================================
# 3. Créer / Charger un modèle pré-entraîné YOLOv8
# ==========================================================
# Tu peux choisir: yolov8n.pt (nano, rapide), yolov8s.pt (small), yolov8m.pt (medium), yolov8l.pt (large)
# Pour un premier entraînement, commence avec YOLOv8n (plus rapide)

model = YOLO("yolov8m.pt")  # charge un modèle pré-entraîné

# ==========================================================
# 4. Lancer l’entraînement
# ==========================================================
model.train(
    data=f"/content/car-parts-detection-1/data.yaml",  # dataset.yaml auto téléchargé
    epochs=50,          # nombre d'époques (ajuste selon ton dataset)
    imgsz=640,          # taille des images
    batch=16,           # taille du batch (ajuste si mémoire insuffisante)          # 0 = GPU, 'cpu' = CPU
    workers=8,          # nombre de workers pour le dataloader
    name="carrosserie_yolov8",  # nom du dossier de sortie
    patience=10,        # early stopping si pas d’amélioration
    optimizer="SGD",    # ou 'AdamW'
    lr0=0.01,           # learning rate initial
    pretrained=True     # commencer avec poids COCO
)

# ==========================================================
# 5. Vérifier les résultats
# ==========================================================
# Les résultats seront enregistrés dans /content/runs/detect/carrosserie_yolov8
# Vérifier les courbes de perte, précision, recall, mAP
```
